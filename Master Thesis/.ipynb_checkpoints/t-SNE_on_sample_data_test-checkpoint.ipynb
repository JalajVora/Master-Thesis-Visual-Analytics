```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
data = np.random.choice(2, size=(40, 20))
```

```python
data
```

```python
df = pd.read_csv('Story_Classification_Data.csv')
```

```python
df
```

```python
t_sne = TSNE(
    n_components=2,
    learning_rate="auto",
    perplexity=20,
    n_iter=5000,
    init="random",
)
projections_tSNE = t_sne.fit_transform(df.iloc[:,:35])
```

```python
projections_tSNE
```

```python
df = pd.DataFrame()
```

```python
transposed_projections = projections_tSNE.T
```

```python
transposed_projections[0]
```

```python
df['X'] = transposed_projections[0]
```

```python
df['Y'] = transposed_projections[1]
```

```python
# plt.figure(figsize=(10,7))
sns.scatterplot(x='X', y='Y', data=df)
```
